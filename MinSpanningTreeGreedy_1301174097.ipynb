```
import networkx as nx
import matplotlib.pyplot as plt
```

```
total = 0
```

```
def shortest(maps, asal, tujuan): # mencari jalur terpendek dari maps (graph yg dibangun)
    result = [] # node dengan jarak terpendek simpan ke dalam list
    result.append(asal) # inisialisasi node pertama dengan nilai asal
    
    global total
    while tujuan not in result: # telusuri graph sampai tujuan ditemukan
        current_node = result[-1] # -1 == list di index terakhir
        print ("Node Terakhir :",current_node)
        jarak_terpendek = min(maps[current_node].values()) # Cari local maximum (nilai/jarak terkecil dari node terakhir ke node selanjutnya)
#        print ("Node Dengan Jarak Terpendek Dari Node Terakhir :|", jarak_terpendek)
       
        for node, jarak in maps[current_node].items(): # iterasi mencari node selanjutnya
            print ("Node Selanjutnya :",node,"| Jarak :",jarak)
            total += jarak_terpendek
            if jarak == jarak_terpendek:
                 # ambil node dengan jarak terpendek dan tambahkan ke list result
                 # agar iterasi selanjutnya dimulai dari node sekarang.
                result.append(node)

    return result
```

```
first_maps = { # graph direpresentasikan dengan menggunakan dictionary di dalam dictionary,
        'A': {'B': 6, 'C': 5},
        'B': {'E': 5, 'F': 8},
        'C': {'B': 5},
        'D': {'E': 6},
        'E': {'F': 4}
       }
```

```
second_maps = { # graph direpresentasikan dengan menggunakan dictionary di dalam dictionary,
        'A': {'B': 8, 'C': 4},
        'B': {'D': 4, 'H': 12,},
        'C': {'D': 10, 'E': 7},
        'D': {'E': 4, 'F': 8, 'G': 10},
        'E': {'F': 3, 'I': 15},
        'F': {'H': 4, 'G': 5},
        'G': {'H': 5, 'I': 10},
        'H': {'I': 7},
       }
```

```
def draw(maps):
    g = nx.DiGraph()
    color = []
    if (maps == int(1)):
        g.add_edge("A", "C", weight="5 km")
        g.add_edge("A", "B", weight="6 km")
        g.add_edge("B", "D", weight="5 km")
        g.add_edge("B", "E", weight="8 km")
        g.add_edge("C", "B", weight="5 km")
        g.add_edge("D", "E", weight="6 km")
        g.add_edge("E", "F", weight="4 km")
        color = ['g','b','b','b','b','r']
    if (maps == int(2)):
        g.add_edge("A", "C", weight="4 km")
        g.add_edge("A", "B", weight="8 km")
        g.add_edge("B", "D", weight="4 km")
        g.add_edge("B", "H", weight="12 km")
        g.add_edge("C", "D", weight="10 km")
        g.add_edge("C", "E", weight="7 km")
        g.add_edge("D", "E", weight="4 km")
        g.add_edge("D", "F", weight="8 km")
        g.add_edge("D", "G", weight="10 km")
        g.add_edge("E", "F", weight="3 km")
        g.add_edge("E", "I", weight="15 km")
        g.add_edge("F", "H", weight="8 km")
        g.add_edge("F", "G", weight="4 km")
        g.add_edge("G", "H", weight="5 km")
        g.add_edge("G", "I", weight="10 km")
        g.add_edge("H", "I", weight="7 km")
        color = ['y','b','b','b','b','b','b','b','r']
    pos = nx.shell_layout(g)
    edge_labels = { (u,v): d['weight'] for u,v,d in g.edges(data=True) }
    nx.draw_networkx_nodes(g,pos,node_size=800, node_color=color)
    nx.draw_networkx_edges(g,pos)
    nx.draw_networkx_labels(g,pos)
    nx.draw_networkx_edge_labels(g,pos,edge_labels=edge_labels)
    plt.title("MAPS")
    plt.axis("off")
    plt.show()
```

```
def _main():
    choose = int(input("Choose Maps :\n1. First Maps\n2. Second Maps\n\n"))
    if (choose == 1):
        print ("\nFirst Maps :")
        draw(choose)
        maps = shortest(first_maps,"A","F")
        print ("\nJalur Dari 'A' ke 'F' :", maps)
    if (choose == 2):
        print ("\nSecond Maps :")
        draw(choose)
        maps = shortest(second_maps,"A","I")
        print ("\nJalur Dari 'A' ke 'I' :", maps)
    
    print('Total Jarak:', total, 'km')
    
if __name__ == '__main__':
    _main()
```
